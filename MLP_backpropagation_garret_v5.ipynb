```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

def criaDataset(n=100, centros=3):
  X, y = make_blobs(n_samples=n, centers=centros, cluster_std=0.2)
  return X, y

X, y = criaDataset()

def plotDataset(X, y):
  plt.xlabel("x1")
  plt.ylabel("x2")
  for k in set(y):
    plt.plot(X[:,0][y==k],X[:,1][y==k], "o", alpha=0.3)

plotDataset(X, y)
```

```python
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.metrics import accuracy_score

def include_bias(X):
  bias = np.ones((X.shape[0], 1))
  Xb = np.concatenate((bias, X), axis=1)
  return Xb

def defineAxes(X):
    offset = 0.5
    min1, max1 = X[:, 0].min(), X[:, 0].max()
    min2, max2 = X[:, 1].min(), X[:, 1].max()
    return [min1-offset, max1+offset, min2-offset, max2+offset]

def plotPredictions(clf, X):
    axes = defineAxes(X)
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
```

```python
from scipy.special import expit
from abc import ABC, abstractmethod

class ActivationFunction(ABC):
  @abstractmethod
  def forward(a):
    pass
  @abstractmethod
  def backward(a):
    pass

class TanH(ActivationFunction):
  def forward(self, a):
    return expit(a)*2-1
  def backward(self, a):
    return (1 - np.square(self.forward(a)))

class Sigmoid(ActivationFunction):
  def forward(self, a):
    return expit(a)
  def backward(self, a):
    return self.forward(a) * (1 - self.forward(a))
```

```python
from sklearn.preprocessing import label_binarize

class Backpropagation(BaseEstimator, ClassifierMixin):
  def __init__(self, max_iter=1000, learning_rate=0.001, n_hidden=[2],
               activation=[TanH()]):
    self.max_iter = max_iter
    self.learning_rate = learning_rate
    self.n_hidden = n_hidden
    self.activation = activation

  def forward(self, X):
    self.A = []
    self.Z = []
    AUX = X.copy()
    for i, W in enumerate(self.Ws):
      self.A.append(include_bias(AUX))
      self.Z.append(self.A[-1] @ W)
      activation = self.activation[min(i, len(self.activation) - 1)]
      AUX = activation.forward(self.Z[-1])
    return AUX

  def backward(self, X, y, ypred):
    grads = []
    output_delta = y - ypred
    grads.insert(0, self.A[-1].T @ output_delta)
    for i in range(len(self.Ws)-1, 0, -1):
      activation = self.activation[min(i, len(self.activation) - 1)]
      grad = activation.backward(self.Z[i-1])
      input_delta = (output_delta @ self.Ws[i][1:,:].T) * grad
      grads.insert(0, self.A[i-1].T @ input_delta)
      output_delta = input_delta.copy()
    for i in range(len(self.Ws)-1, -1, -1):
      self.Ws[i] += grads[i] * self.learning_rate

  def encode_labels(self, y):
    self.labels = list(set(y))
    return label_binarize(y, classes=self.labels)

  def fit(self, X, y):
    yhot = self.encode_labels(y)
    self.Ws = []
    previous_output = X.shape[1]
    for layer_size in self.n_hidden:
      self.Ws.append(np.random.uniform(-1, 1, size=(previous_output+1, layer_size)))
      previous_output = self.Ws[-1].shape[1]
    self.Ws.append(np.random.uniform(-1, 1, size=(previous_output+1, yhot.shape[1])))
    for _ in range(self.max_iter):
      ypred = self.forward(X)
      self.backward(X, yhot, ypred)

  def predict(self, X):
    a = self.forward(X)
    if len(self.labels) <= 2:
      idx = np.array(a > 0, dtype=int).reshape(-1,)
    else:
      idx = np.argmax(a, axis=1)
    ypred = np.array([self.labels[i] for i in idx])
    return ypred

X, y = criaDataset(centros=3)
clf = Backpropagation(n_hidden=[5])
clf.fit(X, y)
plotDataset(X, y)
plotPredictions(clf, X)
ypred = clf.predict(X)
accuracy_score(y, ypred)
```

```python
X, y = criaDataset(centros=10)
clf = Backpropagation(max_iter=1000, learning_rate=0.005, n_hidden=[20, 20], activation=[TanH(),Sigmoid()])
clf.fit(X, y)
plotDataset(X, y)
plotPredictions(clf, X)
ypred = clf.predict(X)
accuracy_score(y, ypred)
```

```python
### Comparação com o sklearn ###

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_validate

X, y = criaDataset(centros=15)

scores = cross_validate(Backpropagation(max_iter=1000, learning_rate=0.005, n_hidden=[20, 20], activation=[TanH(),Sigmoid()]), X, y)
print(f"Custom MLP Scores: {scores['test_score']} {np.mean(scores['test_score']):.2f}")

scores = cross_validate(MLPClassifier(hidden_layer_sizes=(20, 20), max_iter=1000, learning_rate_init=0.005, activation='tanh'), X, y)
print(f"Sklearn MLP Scores: {scores['test_score']} {np.mean(scores['test_score']):.2f}")
```

```python
### Comparação com o sklearn ###

# Configuração inicial e treinamento do modelo Backpropagation
clf = Backpropagation(max_iter=1000, learning_rate=0.005, n_hidden=[20, 20], activation=[TanH(),Sigmoid()])
clf.fit(X, y)
ypred_custom = clf.predict(X)
print("Custom MLP Backprop accuracy:", accuracy_score(y, ypred_custom))

# Treinamento e avaliação do modelo MLPClassifier do sklearn
clf_sklearn = MLPClassifier(hidden_layer_sizes=(20, 20), max_iter=1000, learning_rate_init=0.005, activation='tanh')
clf_sklearn.fit(X, y)
ypred_sklearn = clf_sklearn.predict(X)
print("Sklearn MLP accuracy:", accuracy_score(y, ypred_sklearn))

# Exibindo as previsões dos dois modelos no mesmo gráfico para comparação visual
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plotDataset(X, y)
plotPredictions(clf, X)
plt.title('Custom MLP')

plt.subplot(1, 2, 2)
plotDataset(X, y)
plotPredictions(clf_sklearn, X)
plt.title('Sklearn MLP')
plt.show()
```
